```python
import pandas as pd
df = pd.read_csv('sample_survey_results_public.csv')
cols_needed = ['Hobbyist','Age','ConvertedComp','Country','Employment','LanguageWorkedWith','WorkWeekHrs']
df = df[cols_needed]
```

```python
result = df.groupby('Country').apply(lambda x: pd.Series({
    'NumResponded': len(x),
    'KnowsPython': x['LanguageWorkedWith'].str.contains('Python', na = False).sum(),
    'knowsJavaScript': x['LanguageWorkedWith'].str.contains('JavaScript', na = False).sum(),
    'KnowsJava': x['LanguageWorkedWith'].str.contains('Java', na = False).sum(),
    'ProgrammingAsHobby': (x['Hobbyist'] == 'Yes').sum(),
    'NumStudent': x['Employment'].str.contains('Student', na = False).sum(),
    'AverageAge': x['Age'].mean(),
    'AverageAnnualSalary(USD)': x['ConvertedComp'].mean(),
    'AverageWorkingWeekHours': x['WorkWeekHrs'].mean()
}))

result.to_csv('test.csv')
```
